<h2>Para determinar os valores aproximados de <code>x</code>, <code>y</code> e <code>z</code>, precisamos entender as fórmulas que relacionam os valores de triestímulo (<code>X</code>, <code>Y</code>, <code>Z</code>) e os coeficientes de cromaticidade (<code>x</code>, <code>y</code>, <code>z</code>). Essas fórmulas são derivadas usando as definições do diagrama de cromaticidade CIE 1931.</h2>

<h3>1. Coeficientes de Cromaticidade (<code>x</code>, <code>y</code>, <code>z</code>)</h3>

<p>Os coeficientes de cromaticidade são definidos pelas seguintes fórmulas:</p>

$$
\ x = \frac{X}{X + Y + Z} \
\ y = \frac{Y}{X + Y + Z} \
\ z = \frac{Z}{X + Y + Z} \
$$
<h3>2. Reconstrução dos Valores de Triestímulo (<code>X</code>, <code>Y</code>, <code>Z</code>)</h3>

<p>Dados os coeficientes de cromaticidade e o valor de triestímulo Y, podemos reconstruir os valores de triestímulo usando as seguintes fórmulas:</p>

$$
\ X = \frac{x \cdot Y}{y} \
$$
$$
\ Z = \frac{z \cdot Y}{y} \
$$
<p>Agora, vamos usar essas fórmulas para encontrar os valores aproximados de <code>x</code>, <code>y</code>, <code>z</code>, <code>X</code>, <code>Y</code>, <code>Z</code> com os dados fornecidos:</p>

$$
\ Z = \frac{0.19 \cdot (1.56 + 1.68) }{1 - 0.19} \approx 0.76 \
$$



$$
\ x = \frac{1.68}{1.68 + 1.56 + 0.76} \approx 0.456 \
$$

$$
\ y = \frac{1.56}{1.68 + 1.56 + 0.76} \approx 0.429 \
$$

<p>Portanto, os valores aproximados de triestímulo e coeficientes de cromaticidade são:</p>

$$
\ x \approx 0.42, \quad y \approx 0.39, \quad Z \approx 0.76 \
$$

#**11- Em processamento de imagens coloridas, faça a segmentação do bico da ave na imagem abaixo (pode ser obtida no github: https://github.com/kevinruiz182/image_processing_tools/tree/main/images/color_image_processing)**
Usando as cores no espaço RGB: (R,G,B). Use a distância euclidiana (através da raiz quadrada da soma dos quadrados das diferenças de cada canal de cor respectivo) para determinar a diferença entre as cores dos pixels da imagem e o pixel amarelo, determinado por (255,255,0). Considere cada pixel como um vetor e o range de cada canal sendo de 0 a 255.  Seja a distância de limiar a mesma distância entre os pixels: (0,0,0) e (100,100,100).


```python
import numpy as np
import cv2

BALD_EAGLE_PATH = "/content/drive/MyDrive/data/bald_eagle_portrait.jpg"
im = cv2.imread(BALD_EAGLE_PATH)

cv2_imshow(im)
```

```python
thr = np.sqrt(np.power(100,2)*3)
print('Threshold', thr)

yellow_pixel = np.array([255,255,0], dtype=np.float32)

list_dist = []
for row in range(im.shape[0]):
    for col in range(im.shape[1]):
        R, G, B = im[row,col,2], im[row,col,1], im[row,col,0]
        pixel = np.array([R,G,B], dtype=np.float32)
        diff = np.linalg.norm(pixel-yellow_pixel)
        if diff < thr:
            list_dist.append(diff)


print('Mean dist', np.mean(list_dist))
```

```python
MOON_PATH = '/content/drive/MyDrive/data/moon.jpg'
image = cv2.imread(MOON_PATH)
cv2_imshow(image)
```

```python

# Somar os 3 canais da imagem original
sum_channels = np.sum(image, axis=2)

# Normalizar o resultado pela função fornecida e dividir por 255
normalized_image = (255 * (sum_channels - sum_channels.min()) / (sum_channels.max() - sum_channels.min())).astype(int) / 255.0

# Aplicar a aproximação conforme as condições dadas
output_image = np.zeros_like(normalized_image)
output_image[normalized_image < 0.25] = 0
output_image[(normalized_image >= 0.25) & (normalized_image < 0.5)] = 0.25
output_image[(normalized_image >= 0.5) & (normalized_image < 0.75)] = 0.5
output_image[normalized_image >= 0.75] = 1.0

# Exibir a imagem resultante
plt.imshow(output_image, cmap='gray')
plt.title("Imagem Resultante com Segmentação")
plt.show()

# Calcular a média dos pixels sobre toda a imagem
average_value = np.mean(output_image)
print(f"A média dos pixels sobre toda a imagem é: {average_value}")
```
